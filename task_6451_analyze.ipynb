```python
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
df = pd.read_csv('hmc_database/merged_ext_train.csv', index_col=0)
df.tail(10)
```

```python
df = df[df['sum'] != 0]
checks_df = df[['check_id', 'sum', 'datetime']].drop_duplicates()
print('chekcs_count: ', len(checks_df))
checks_df.sort_values(by=['datetime'], ignore_index=True, inplace=True)
checks_df
```

```python
checks_df.describe()['sum']
```

```python
max_date = pd.to_datetime(checks_df.datetime.max())
min_date = pd.to_datetime(checks_df.datetime.min())
print('first day: ', min_date)
print('last day: ', max_date)
print('diff between first and last day in months: ', len(pd.date_range(start=min_date, end=max_date, freq='M')))
print('diff between first and last day in days: ', len(pd.date_range(start=min_date, end=max_date, freq='D')))
print('diff between first and last day in weeks: ', len(pd.date_range(start=min_date, end=max_date, freq='W')))
```

```python
px.scatter(checks_df, x=checks_df['datetime'], y=checks_df['sum'], title='check_sum_by_date')
```

```python
checks_df['month'] = pd.to_datetime(checks_df['datetime']).dt.month
months_df = checks_df.groupby('month').describe()
months_df
```

```python
px.scatter(x=months_df.index.to_numpy(), y=months_df['check_id', 'count'].to_numpy(), title='check count per month')
```

```python
px.scatter(x=months_df.index.to_numpy(), y=months_df['sum', 'mean'], title='avg sum per month')
```

```python
check_month_sum = checks_df.groupby('month')['sum'].sum()
px.scatter(check_month_sum, x=check_month_sum.index, y=check_month_sum, title='sum count per month')
```

```python
checks_df['day'] = pd.to_datetime(checks_df['datetime']).dt.day
day_df = checks_df.drop(columns='month').groupby('day').describe()
day_df
```

```python
px.scatter(x=day_df.index.to_numpy(), y=day_df['check_id', 'count'].to_numpy(), title='check count per day')
```

```python
px.scatter(x=day_df.index.to_numpy(), y=day_df['sum', 'mean'], title='avg sum per day')
```

```python
check_day_sum = checks_df.groupby('day')['sum'].sum()
px.scatter(check_day_sum, x=check_day_sum.index, y=check_day_sum, title='sum count per day')
```
