```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('hmc_database/bot-log-2 - data.csv')
df.head()
```

```python
df.columns
```

```python
df = df.dropna(subset=['категория'], axis=0)
df.head()
```

```python
print('inn len:', len(pd.unique(df['seller-inn'])
                      ))

print('df len: ', len(df))
```

```python
inn_category_series = df.groupby('seller-inn').категория.unique()
inn_category_series
```

```python
print("len: ", len(inn_category_series))
inn_len_count = [0, 0]
for inn, category_list in inn_category_series.items():
    print('inn: ', inn)
    inn_len = len(str(inn))
    if inn_len == 10:
        inn_len_count[0] += 1
    elif inn_len == 12:
        inn_len_count[1] += 1
    print("category_list: \n", category_list)
print("Юридические лица(инн длины 10)", inn_len_count[0])
print("Физические лица(инн длины 12)", inn_len_count[1])
# инн юридического(физического) является последовательностью из 10(12) цифр, из которых первые 4 цифры
# — код налогового органа, который присвоил ИНН,
# следующие 6(5) —
# порядковый номер записи о налогоплательщике в территориальном разделе
# Единого государственного реестра налогоплательщиков налогового органа,
# который присвоил ИНН,
# и последние 2(1) — так называемое «контрольное число» для проверки правильности номера.
#
#
```

```python
print('юридические')
uri_all_category = []
for inn, category_list in inn_category_series.items():
    inn_len = len(str(inn))
    if inn_len == 10:
        print('inn: ', inn)
        print("category_list: \n", category_list)
        uri_all_category += list(category_list)
uri_all_category = list(set(uri_all_category))
```

```python
print('физические')
ph_all_category = []
for inn, category_list in inn_category_series.items():
    inn_len = len(str(inn))
    if inn_len == 10:
        print('inn: ', inn)
        print("category_list: \n", category_list)
        ph_all_category += list(category_list)
ph_all_category = list(set(ph_all_category))
```

```python
print('physic_info: ')
print('len', len(ph_all_category))
for ph in ph_all_category:
    print(ph)
print('////\n\n\n/////')
print('uri_info:')
print('len', len(uri_all_category))
for uri in uri_all_category:
    print(uri)

###юридические и физические дают одинаковые категории
```
