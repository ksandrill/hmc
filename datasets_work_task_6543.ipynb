```python
import pandas as pd
```

```python
train_df = pd.read_csv('hmc_database/train.csv')
train_df.head(10)
```

```python
train_df['category'].unique()
```

```python
train_checks_df = pd.read_csv('hmc_database/train_checks.csv')
train_checks_df.head(10)
```

```python
merge_df = pd.merge(train_df, train_checks_df, how='left', on='check_id')
merge_df.sort_values(by=['datetime', 'check_id'], ignore_index=True, inplace=True)
merge_df.head(20)
```

```python
merge_df.to_csv('hmc_database/merged_ext_train.csv')
```
